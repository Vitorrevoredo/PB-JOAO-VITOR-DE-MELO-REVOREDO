```python
arquivo = open('./actors.csv', 'r')
linhas= arquivo.readlines()
arquivo.close()
for c in linhas:
    print(c)
```

# Etapa 1
Apresentar o ator/atriz com maior numero de filmes e a respectiva quantidade. A quantidade de filmes encontrados encontra-se na coluna Number of movies do arquivo

```python
maiorNumeroFilmes = 0
atorMaisFilmes = ""

for linha in linhas[1:]:
    if '"' in linha:
        partes = linha.split('"')
        nome_ator = partes[1]
        resto_dados = partes[2].split(',')
        numero_filmes = int(resto_dados[2].strip())
    else:
        partes = linha.split(',')
        nome_ator = partes[0].strip()
        numero_filmes = int(partes[2].strip())

    if numero_filmes > maiorNumeroFilmes:
        maiorNumeroFilmes = numero_filmes
        atorMaisFilmes = nome_ator

with open('etapa-1.txt', 'w') as output:
    output.write(f" {atorMaisFilmes} com {maiorNumeroFilmes} filmes.")
```

# Etapa 2
Apresente a média de receita de bilheteria bruta dos principais filmes, considerando todos os atores. Estamos falando aqui da media da coluna Gross

```python
cabecalho = linhas[0]
dados = linhas[1:]

soma_receitas = 0
contador = 0

for linha in dados:
    if '"' in linha:
        partes = linha.split('"')
        resto_dados = partes[2].split(',')
        receita = resto_dados[5].strip()
    else:
        parte = linha.split(',')
        receita = parte[5].strip()

    if receita:
        try:
            soma_receitas += float(receita)
            contador += 1
        except ValueError:
            continue

if contador > 0:
    media_receita = soma_receitas / contador
else:
    media_receita = 0

with open('etapa-2.txt', 'w') as arquivo_saida:
    arquivo_saida.write(f'Media de receita bruta dos principais filmes: {media_receita:.2f}\n')
```

# Etapa 3 
Apresente o ator/atriz com maior média de receita de bilheteria bruta por filme do conjunto de dados. Considere a coluna Average per Movie para fim de calculo

```python
cabecalho = linhas[0]
dados = linhas[1:]

maior_media_receita = 0
ator_maior_media = ""

for linha in dados:
    if '"' in linha:
        partes = linha.split('"')
        ator = partes[1]
        resto_dados = partes[2].split(',')
        media_receita = resto_dados[3].strip()
    else:
        parte = linha.split(',')
        ator = parte[0].strip()
        media_receita = parte[3].strip()  

    if media_receita: 
        try:
            media_receita_float = float(media_receita)
            if media_receita_float > maior_media_receita:
                maior_media_receita = media_receita_float
                ator_maior_media = ator
        except ValueError:
            continue 

with open('etapa-3.txt', 'w') as arquivo_saida:
    arquivo_saida.write(f'{ator_maior_media} com uma media de {maior_media_receita:.2f}\n')
```

# Etapa 4 
A coluna #1 Movie contém o filme de maior bilheteria em que o ator atuou. Realize a contagem de aparições destes filmes no dataset, listando-os pela quantidade de vezes que estão presentes. Considere a ordem decrescente e, em segundo nível, o nome do filme.
Ao escrever no arquivo, considere o padrão de saída (sequencia) - O filme (nome filme) aparece (quantidade) vez(es) no dataset, adicionando um resultado a cada linha.

```python
dados = linhas[1:]
contagem_filmes = {}

for linha in dados:
        parte = linha.split(',')
        filme = parte[4].strip()
        if filme:
            if filme in contagem_filmes:
                contagem_filmes[filme] += 1
            else:
                contagem_filmes[filme] = 1

filmes_ordenados = sorted(contagem_filmes.items(), key=lambda x: (x[1], x[0]), reverse=True)
resultados = []
for filme, contagem in filmes_ordenados:
    resultados.append(f'O filme "{filme}" aparece {contagem} vez(es) no dataset.')

with open('etapa-4.txt', 'w') as arquivo_saida:
    arquivo_saida.write('\n'.join(resultados) + '\n')
```

# Etapa 5
Apresente a lista dos atores ordenada pela receita bruta de bilheteria de seus filmes(coluna Total Gross), em ordem decrescente. Ao escrever no arquivo, considere o padrão de saída (nome do ator)  (receita total bruta), adicionando um resultado a cada linha

```python
arquivo = open('./actors.csv', 'r')
linhas = arquivo.readlines()
arquivo.close()
cabecalho = linhas[0]
dados = linhas[1:]
receitas_atores = {}

for linha in dados:
    if '"' in linha:
        partes = linha.split('"')
        ator = partes[1]
        resto_dados = partes[2].split(',')
        total_gross = resto_dados[5].strip()
    else:
        parte = linha.split(',')
        ator = parte[0].strip()
        total_gross = parte[5].strip()
    
    if total_gross: 
        try:
            total_gross_float = float(total_gross)
            if ator in receitas_atores:
                receitas_atores[ator] += total_gross_float
            else:
                receitas_atores[ator] = total_gross_float
        except ValueError:
            continue 

atores_ordenados = sorted(receitas_atores.items(), key=lambda x: x[1], reverse=True)

resultados = []
for ator, receita in atores_ordenados:
    resultados.append(f'{ator} ({receita:.2f})')

with open('etapa-5.txt', 'w') as arquivo_saida:
    arquivo_saida.write('\n'.join(resultados) + '\n')
```

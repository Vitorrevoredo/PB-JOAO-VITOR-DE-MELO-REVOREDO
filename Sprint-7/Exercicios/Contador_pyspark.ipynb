```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("ContagemDePalavras").getOrCreate()
```

```python
arquivo_texto = spark.sparkContext.textFile("README.md")
```

```python
# Dividindo o conte√∫do em palavras
palavras = arquivo_texto.flatMap(lambda linha: linha.split())

# Contando as ocorr√™ncias de cada palavra
contagem_palavras = palavras.map(lambda palavra: (palavra, 1)).reduceByKey(lambda a, b: a + b)
```

```python
contagem_palavras.collect()
```

```python
total_ocorrencias = contagem_palavras.map(lambda x: x[1]).sum()
```

```python
print(f"Total de ocorr√™ncias de palavras: {total_ocorrencias}")
```
